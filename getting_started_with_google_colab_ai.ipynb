```python
import pandas as pd
dados_vendas ={'id_pedido':[101,102,103,104,105],
               'produto':['bag sport premium','Sneaker Horizon','Bag Sport premium','caps limited', 'Sneaker Horizon'],
               'quantidade_vendidas':[2,1,1,5,2]
               }
df_vendas= pd.DataFrame(dados_vendas)
dados_produtos = {
    'produto': ['bag sport premium', 'Sneaker Horizon','T-shirt Peerformace','caps limited'],
    'preco_Unitario': [250,450,120,80],
    'custo_unitario': [150,250,70,40]
}
df_produtos=pd.DataFrame(dados_produtos)

print("___TABELA DE VENDAS___")
display(df_vendas)
print("/n___TABELA DE PRODUTOS (Financeiros)___")
display(df_produtos)
```

```python
# Cruzando as tabelas: Juntamos df_vendas com df_produtos usando a coluna 'Produto' como ponte
df_consolidado = pd.merge(df_vendas, df_produtos, on='produto', how='inner')

# Exibindo o resultado final cruzado
df_consolidado
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Criamos os dados dos meses (1 a 5) e o faturamento real que aconteceu
meses = np.array([1, 2, 3, 4, 5])
faturamento = np.array([10000, 15000, 22000, 28000, 35000])

# 2. A mágica matemática: o Python calcula a linha de tendência (coeficientes)
modelo = np.polyfit(meses, faturamento, 1)
funcao_previsao = np.poly1d(modelo)

print("Padrão de crescimento mapeado com sucesso!")
```

```python
# 3. Definimos o período completo (o passado + o futuro que queremos prever)
meses_futuros = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# 4. O Python aplica a fórmula matemática para prever os valores
faturamento_predito = funcao_previsao(meses_futuros)

# 5. Desenhando o gráfico de linha estratégico
plt.figure(figsize=(8, 5))
plt.plot(meses, faturamento, 'go', label='Faturamento Real (Passado)') # Pontos verdes
plt.plot(meses_futuros, faturamento_predito, 'r--', label='Linha de Tendência (Previsão)') # Linha tracejada vermelha

plt.title('Previsão de Faturamento para os Próximos Meses')
plt.xlabel('Mês')
plt.ylabel('Faturamento (R$)')
plt.legend()
plt.grid(True)
plt.show()

# Printando os valores exatos da previsão na tela
print(f"Previsão para o Mês 6: R$ {funcao_previsao(6):.2f}")
print(f"Previsão para o Mês 7: R$ {funcao_previsao(7):.2f}")
print(f"Previsão para o Mês 8: R$ {funcao_previsao(8):.2f}")
```

```python
# Crie a coluna Faturamento_Total (multiplicando a Quantidade_Vendida pelo Preco_Unitario).
df_consolidado['Faturamento_Total'] = df_consolidado['quantidade_vendidas'] * df_consolidado['preco_Unitario']

# Crie a coluna Lucro_Total (multiplicando a Quantidade_Vendida pela diferença entre Preco_Unitario e Custo_Unitario).
df_consolidado['Lucro_Total'] = df_consolidado['quantidade_vendidas'] * (df_consolidado['preco_Unitario'] - df_consolidado['custo_unitario'])

# Exibindo o resultado final cruzado com as novas colunas
display(df_consolidado)
```

```python
def analisar_mes(faturamento,custo):
    lucro=faturamento-custo
    if lucro>0:
        return f"lucro de R${lucro}:)"
    elif lucro==0:
        return "empatou!R$0"
    else:
        return f"R${lucro}de PREJUIZO:("
historico_vendas=[1500,1200,9000,20000]
custo_fixo= 12000
print("___RELATORIO DE DESEMPENHO MENSAL___")
for faturamento_do_mes in historico_vendas:
  resultado = analisar_mes(faturamento_do_mes,custo_fixo)
  print(f"faturamento:R${faturamento_do_mes}->resultado:{resultado}")
```

For longer text generations, you can stream the response. This displays the output token by token as it's generated, rather than waiting for the entire response to complete. This provides a more interactive and responsive experience. To enable this, simply set stream=True.

```python
# @title Simple streaming example
from google.colab import ai

stream = ai.generate_text("Tell me a short story.", stream=True)
for text in stream:
  print(text, end='')
```

```python
#@title Text formatting setup
#code is not necessary for colab.ai, but is useful in fomatting text chunks
import sys

class LineWrapper:
    def __init__(self, max_length=80):
        self.max_length = max_length
        self.current_line_length = 0

    def print(self, text_chunk):
        i = 0
        n = len(text_chunk)
        while i < n:
            start_index = i
            while i < n and text_chunk[i] not in ' \n': # Find end of word
                i += 1
            current_word = text_chunk[start_index:i]

            delimiter = ""
            if i < n: # If not end of chunk, we found a delimiter
                delimiter = text_chunk[i]
                i += 1 # Consume delimiter

            if current_word:
                needs_leading_space = (self.current_line_length > 0)

                # Case 1: Word itself is too long for a line (must be broken)
                if len(current_word) > self.max_length:
                    if needs_leading_space: # Newline if current line has content
                        sys.stdout.write('\n')
                        self.current_line_length = 0
                    for char_val in current_word: # Break the long word
                        if self.current_line_length >= self.max_length:
                            sys.stdout.write('\n')
                            self.current_line_length = 0
                        sys.stdout.write(char_val)
                        self.current_line_length += 1
                # Case 2: Word doesn't fit on current line (print on new line)
                elif self.current_line_length + (1 if needs_leading_space else 0) + len(current_word) > self.max_length:
                    sys.stdout.write('\n')
                    sys.stdout.write(current_word)
                    self.current_line_length = len(current_word)
                # Case 3: Word fits on current line
                else:
                    if needs_leading_space:
                        # Define punctuation that should not have a leading space
                        # when they form an entire "word" (token) following another word.
                        no_leading_space_punctuation = {
                            ",", ".", ";", ":", "!", "?",        # Standard sentence punctuation
                            ")", "]", "}",                     # Closing brackets
                            "'s", "'S", "'re", "'RE", "'ve", "'VE", # Common contractions
                            "'m", "'M", "'ll", "'LL", "'d", "'D",
                            "n't", "N'T",
                            "...", "…"                          # Ellipses
                        }
                        if current_word not in no_leading_space_punctuation:
                            sys.stdout.write(' ')
                            self.current_line_length += 1
                    sys.stdout.write(current_word)
                    self.current_line_length += len(current_word)

            if delimiter == '\n':
                sys.stdout.write('\n')
                self.current_line_length = 0
            elif delimiter == ' ':
                # If line is full and a space delimiter arrives, it implies a wrap.
                if self.current_line_length >= self.max_length:
                    sys.stdout.write('\n')
                    self.current_line_length = 0

        sys.stdout.flush()
```

```python
# @title Formatted streaming example
from google.colab import ai

wrapper = LineWrapper()
for chunk in ai.generate_text('Give me a long winded description about the evolution of the Roman Empire.', model_name='google/gemini-2.0-flash', stream=True):
  wrapper.print(chunk)
```
